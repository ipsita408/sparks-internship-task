# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv('SampleSuperstore.csv')
data
```

```python
# as country remains constant that is United states in all cases we can drop that features coz that won't be a factor affecting.
data.drop('Country',axis=1, inplace = True)
data
```

```python
data.drop_duplicates()
```

# Correlation matrix 

```python
from matplotlib import style
style.use("seaborn-deep")
corr=data.corr()
sns.heatmap(corr,annot=True)
```

# Pairplot for shipmode 

```python
sns.pairplot(data,hue="Ship Mode")
```

```python
data['Ship Mode'].value_counts()
```

```python
sns.countplot(x=data['Ship Mode'])
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Ship Mode",y="Profit",data=data)
plt.show()
```

## Pair plot for category

```python
sns.pairplot(data,hue="Category")
```

```python
data["Category"].value_counts()
```

```python
sns.countplot(x="Category",data=data)
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Category",y="Profit", data=data)
plt.show()
```

## plot for sub-category

```python
data['Sub-Category'].value_counts()
```

```python
sns.pairplot(data,hue="Sub-Category")
```

# plot for states

```python
plt.figure(figsize=(15,12))
sns.countplot(x="State",data=data,palette="rocket",order=data['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
# profit vs sales and quantity
plt.scatter(data['Sales'],data['Profit'])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

```python
sns.barplot(x="Region",y="Sales",hue="Segment",data=data,order=data['Region'].value_counts().index,palette='rocket')
plt.xlabel("Region",fontsize=10)
plt.show()
```

```python
data.groupby('Region')[["Profit","Sales"]].sum().plot.bar(color=["green","red"],figsize=(8,5))
plt.ylabel("profit/loss and sales")
plt.show()
```

```python
data.groupby('State')[["Profit","Sales"]].sum().sort_values(by="Sales",ascending=False).plot.bar(color=["green","red"],figsize=(8,5))
plt.ylabel("profit/loss and sales")
plt.xlabel("states")
plt.ylabel("profit/loss and sales")
plt.title
plt.show()
```

```python
data.groupby('Category')[["Profit","Sales"]].sum().plot.bar(color=["green","red"],figsize=(8,5))
plt.ylabel("profit/loss and sales")
plt.show()
```

```python
data.groupby('Sub-Category')[["Profit","Sales"]].sum().sort_values(by="Sales").plot.bar(color=["green","red"],figsize=(8,5))
plt.ylabel("profit/loss and sales")
plt.show()
```
