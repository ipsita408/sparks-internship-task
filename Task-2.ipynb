# PREDICTION USING UNSUPERVISED LEARNING ML

In this section from the Data set"Iris" we will predict the optimum no. of clusters and represnt it visually.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

```python
data= pd.read_csv("Iris.csv")
data
```

```python
data_mapped= data.copy()
data_mapped['Species']=data_mapped['Species'].map({'Iris-setosa':0,'Iris-virginica':2,'Iris-versicolor':1})
data_mapped
```

## selecting features

```python
x=data_mapped.iloc[:,0:6]
x
```

```python
kMeans=KMeans(2) 
kMeans.fit(x)
identified_clusters = kMeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters=data_mapped.copy()
data_with_clusters['cluster']=identified_clusters
data_with_clusters
```

 # Selecting appropriate Number of clusters (The Elbow Method)

```python
kMeans.inertia_
```

```python
wcss=[]

for i in range(1,11):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
    
wcss
```

```python
number_clusters = range(1,11)
plt.plot(number_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

from the above elbow shapped graph we can see that there are 2 sharp points one is at 2 and other is 3 and the optimum no, of clusters is where the elbow occurs and 3 would be more apppropriate and better clustering .

```python
kMeans=KMeans(3) 
kMeans.fit(x)
y_clusters = kMeans.fit_predict(x)
y_clusters
```

```python
data_with_clusters=data_mapped.copy()
data_with_clusters['cluster']=y_clusters
data_with_clusters
```

## visualization through scatter plot

```python
plt.scatter(data_with_clusters['SepalLengthCm'],data_with_clusters['SepalWidthCm'],c=data_with_clusters['cluster'],cmap='rainbow')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
# plotting centriods
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')
plt.show()
```
