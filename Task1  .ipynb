# PREDICTION USING SUPERVISED ML

In this section we will use stats model for regression analysis . We will start with simple linear regression involving two variables.


In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.



## Importing Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

## load the data

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
data
```

```python
y=data["Scores"]
x1=data["Hours"]
```

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

from the above regression results we know that all the variables are significant as the P-values are less than 0.05. 

We also get the value of constants for the linear equation that is y=bo+x1b1 and the F-statistic value is also high so our model is pretty good.

## Visualization and Regression line 

```python
plt.scatter(x1,y)
yhat=9.7758*x1+2.4837
plt.plot(x1,yhat,lw=3,c="r",label="regression line")
plt.xlabel('Hours studied',fontsize=15)
plt.ylabel('Scores obtained',fontsize=15)
plt.legend()
plt.show()
```

## predictions:-

```python
data_predicted=results.predict()
data_predicted
```

```python
df=pd.DataFrame({'Actual':data["Scores"],'predicted':data_predicted})
df
```
